```python
#data visualisation
#matplotlib
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv(r"C:\Users\janan\OneDrive\Desktop\Internship\py git\Student_dataset_1.csv")
```

```python
df.plot(x='DEPARTMENT', y='ARREARS', kind='line', marker='o')
plt.xlabel('ColumnX')    
plt.ylabel('ColumnY')   
plt.title('Line Plot of ColumnY vs ColumnX')
plt.show()
```

```python
df.plot(x='PERCENT', y='PASS/FAIL', kind='scatter', color='green')
plt.xlabel('percent')
plt.ylabel('pass or fail')
plt.title('Scatter Plot')
plt.show()
```

```python
# aggregate by dept and plot arrears
df_grouped = df.groupby('DEPARTMENT')['ARREARS'].mean()
df_grouped.plot(kind='bar')
plt.title('Arrears each dept')
plt.xlabel('Department')
plt.ylabel('Arrears')
plt.show()
```

```python
df['ARREARS'].plot(kind='hist', bins=10, edgecolor='black', color='violet')
plt.xlabel('NumericColumn')
plt.title('Histogram of arrears')
plt.show()
```

```python
import pandas as pd

df_grouped = df.groupby(['DEPARTMENT', 'GENDER'])['ARREARS'].mean().reset_index()
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10,6))

sns.barplot(data=df_grouped, x='DEPARTMENT', y='ARREARS', hue='GENDER')

plt.title('Average Marks by Department and Gender')
plt.xlabel('Department')
plt.ylabel('Average Marks')
plt.legend(title='Gender')
plt.show()
```
